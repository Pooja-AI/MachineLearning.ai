### Classification

<font size="3"><p>Classification is a type of supervised learning. It specifies the class to which data elements belong to and is best used when the output has finite and discrete values. It predicts a class for an input variable as well. </p><br>
There are 2 types of Classification: <br>
    <ul><li>Binomial</li>
        <li>Multi-Class</li></ul> </font>
